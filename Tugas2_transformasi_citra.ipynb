1.Transformasi Linier (Negative dan
Ident

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar
img = cv2.imread("/content/istockphoto-102723363-612x612-1.jpg")

# Konversi BGR ke RGB agar benar saat ditampilkan
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Transformasi Identity (gambar tetap sama)
identity = img.copy()

# Transformasi Negatif
negative = 255 - img

# Menampilkan hasil
plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("Gambar Asli")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(identity)
plt.title("Transformasi Identity")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(negative)
plt.title("Transformasi Negatif")
plt.axis("off")

plt.tight_layout()
plt.show()
```

2.Transformasi Logaritmik (Log dan
Inverse

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar
img = cv2.imread("/content/images.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Normalisasi ke 0-1
img_norm = img / 255.0

# ===== Log Transformation =====
c = 1
log_transform = c * np.log(1 + img_norm)

# Normalisasi kembali ke 0-255
log_transform = np.uint8(255 * log_transform / np.max(log_transform))

# ===== Inverse Log Transformation =====
inv_log = c * (np.exp(img_norm) - 1)

# Normalisasi kembali
inv_log = np.uint8(255 * inv_log / np.max(inv_log))

# ===== Menampilkan hasil =====
plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("Gambar Asli (RGB)")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(log_transform)
plt.title("Log Transform")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(inv_log)
plt.title("Inverse Log Transform")
plt.axis("off")

plt.tight_layout()
plt.show()
```

3.Transformasi Power-Law (nth power dan nth root power)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("/content/istockphoto-102723363-612x612-1.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Kernel sharpening
kernel = np.array([[0,-1,0],
                   [-1,5,-1],
                   [0,-1,0]])

sharp = cv2.filter2D(img, -1, kernel)

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Gambar Asli")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(sharp)
plt.title("Sharpening")
plt.axis("off")

plt.show()
```
